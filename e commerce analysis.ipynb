```python
#by using this loop we able to display the complete data set as file
tab_size=20
for i in range(0,len(df),tab_size):
    display(df[i:i+tab_size])
```

```python
df['Order_date']=pd.to_datetime(df['Order_date'])
df['Month_year']=df['Order_date'].apply(lambda x : x.strftime('%Y-%m'))
print(df['Month_year'])
```

```python
df.head()
```

```python
df.groupby('Month_year').count().reset_index()
```

```python
# grouping the data set based on month year 
sales_trend = df.groupby('Month_year').sum('Sales').reset_index()
display(sales_trend)
#creating the graph based on monthly sales
plt.figure(figsize=(12,6))
plt.title('Monthly Sales')
bar1=plt.bar(sales_trend['Month_year'],sales_trend['Sales'])
for bars in bar1:
    height=bars.get_height()
    plt.text(bars.get_x() + bars.get_width() / 2, height, f'{height}', ha='center', va='bottom')
plt.show()
```

```python
#finding the top three months sales and representing in graph
top_three_months=sales_trend.sort_values(by='Sales',ascending=False).head(3)
plt.figure(figsize=(4,5))
plt.title('top three months sales')
bar2=plt.bar(top_three_months['Month_year'],top_three_months['Sales'])
for bars in bar2:
    height = bars.get_height()
    plt.text(bars.get_x() + bars.get_width() / 2, height, f'{height}', ha='center', va='bottom')
plt.show()
plt.show()
```

```python
# grouping the data based on category
category_trend=df.groupby('Category',as_index=False).sum('Sales')
display(category_trend)
```

```python
#figure based on the category and its sales
plt.figure(figsize=(20,10))
plt.plot(category_trend['Category'],category_trend['Sales'],color='red')
bar3=plt.bar(category_trend['Category'],category_trend['Sales'])
plt.title('Category wise sales')
for bars in bar3:
    height = bars.get_height()
    plt.text(bars.get_x() + bars.get_width() / 2, height, f'{height}', ha='center', va='bottom')
plt.xticks(rotation=90)
plt.show()
```

```python
#finding the category wise stock quantity
plt.figure(figsize=(20,10))
plt.plot(category_trend['Category'],category_trend['StockQuantity'],color='red')
bar3=plt.bar(category_trend['Category'],category_trend['StockQuantity'])
plt.title('Category wise sales')
for bars in bar3:
    height = bars.get_height()
    plt.text(bars.get_x() + bars.get_width() / 2, height, f'{height}', ha='center', va='bottom')
plt.xticks(rotation=90)
plt.show()
```

```python
#finding the top 3 category sales
top_three=category_trend.sort_values(by='Sales',ascending=False).head(3)
plt.figure(figsize=(4,5))
plt.title('top three category sales')
bar4=plt.bar(top_three['Category'],top_three['Sales'])
for bars in bar4:
    height = bars.get_height()
    plt.text(bars.get_x() + bars.get_width() / 2, height, f'{height}', ha='center', va='bottom')
plt.xticks(rotation=90)
plt.show()
```

```python
#finding least 3 category sales
least_three=category_trend.sort_values(by='Sales',ascending=False).tail(3)
plt.figure(figsize=(4,5))
plt.title('least three category sales')
bar4=plt.bar(least_three['Category'],least_three['Sales'])
for bars in bar4:
    height = bars.get_height()
    plt.text(bars.get_x() + bars.get_width() / 2, height, f'{height}', ha='center', va='bottom')
plt.xticks(rotation=90)
plt.show()
```
